```python
from keras.layers import Input,Dense,Embedding,LSTM,Dropout,TimeDistributed,Bidirectional
from keras.models import Model,load_model
from keras.utils import np_utils
import numpy as np
import re
```

```python
inputfile='train_cws.txt'
vocab=open(inputfile,encoding='utf-8').read().rstrip('\n').split('\n')
```

```python
#判断是否为中文字符
def is_Chinese(word):
    for ch in word:
        if '\u4e00' <= ch <= '\u9fff':
            return True
    return False
  
#统计词频字典
#newvocab=list(''.join(vocab))
word_freq={}
for s in vocab:
    for word in s.split():
        if is_Chinese(word):
            for i in word:
                word_freq[i]=word_freq.get(i,0)+1
        else:
            word_freq[word]=word_freq.get(word,0)+1
word_freq
```

```python
#按照词频顺序对字典逆序排序
stat=sorted(word_freq.items(),key=lambda x:x[1],reverse=True)

#为每个字编码索引号
words=[s[0] for s in stat]
print(len(words))
char2id={c:i+1 for i,c in enumerate(words)}
id2char={i+1:c for i,c in enumerate(words)}
len(char2id)   
```

```python
#字的标签
tags={'s':0,'b':1,'m':2,'e':3,'x':4}
#定义模型所需的参数
embedding_size=128  #字嵌入的长度
maxlen=0   #长于32则截断，短于32则填充0
hidden_size=64
batch_size=64
epochs=50
```

```python
#将数据集切分成短小的句子
data=open(inputfile,encoding='utf-8').read().rstrip('\n')
data = re.split('[，。！、？\n]', data)
print(len(data))
sentences=[]
for i in data:
    if len(i)>0:
        sentences.append(i)
max(len(d.split()) for d in data)
```

```python
from collections import OrderedDict #维护字典有序插入
#准备数据
X_data=[]
y_data=[]
dic = OrderedDict()
for sentence in sentences:
    sentence=sentence.split(' ')
    X=[]
    y=[]
    try:
      for word in sentence:
          word=word.strip()
          #print(word)
          if is_Chinese(word):
              if len(word)==0:
                  continue
              elif len(word)==1:
                  X.append(char2id[word])
                  y.append(tags['s'])
                  dic[word]='s'
              elif len(word)>1:
                  X.append(char2id[word[0]])
                  y.append(tags['b'])
                  dic[word[0]]='b'
                  for i in range(1,len(word)-1):
                      X.append(char2id[word[i]])
                      y.append(tags['m'])
                      dic[word[i]]='m'
                  X.append(char2id[word[-1]])
                  y.append(tags['e'])   
                  dic[word[-1]]='e'
          else:
              X.append(char2id[word])
              y.append(tags['s'])
              dic[word]='s'
    except:
        continue
    else:
        if len(X)>0:
          X_data.append(X)
          y_data.append(y)
X_data
```
